```python
#naive imputation by the precedent value
def imputation_precedent(data):
    m,n = data.shape
    for i in range(m):
        for j in range(n):
            if data[i,j] == -1:
                data[i,j] = data[i,j-1]
    return None

imputation_precedent(datas)

#datas = datas.transpose()  # data in colomn or in line?
```

```python
from sklearn.preprocessing import MinMaxScaler

#Feature Scaling

scaler = MinMaxScaler()                                                 
datas_scale = scaler.fit_transform(datas.transpose()) # MinMaxScaler applied to Columns
datas_scale = datas_scale.transpose()



#sampling of p% of datasets and creation of label
np.random.seed(42)
p=0.1
sample_dim1 = np.random.choice(range(datas.shape[0]),int(np.sqrt(p)*datas.shape[0]),replace= True)
sample_dim2 = np.random.choice(range(datas.shape[1]),int(np.sqrt(p)*datas.shape[1]),replace= True)

label = np.zeros(datas.shape)
print(sample_dim1.shape,sample_dim2.shape, label.shape)

#adding gaussien noise to p% of the dataset, mu mean, sigma ecart-type

mu = 1.5 * np.mean(datas_scale)
sigma = 0.2*np.mean(datas_scale)

for i in range(sample_dim1.size):
    for j in range(sample_dim2.size):
        label[sample_dim1[i],sample_dim2[j]] = 1
        datas_scale[sample_dim1[i],sample_dim2[j]] += np.random.normal(mu,sigma)

 # verifacation whether we have p% datas
print(label,label.shape)
print(datas_scale,datas_scale.shape)

```

```python
import matplotlib.pyplot as plt
shuffled_indices = np.random.permutation(datas.shape[0])
visu_sens_size = shuffled_indices[:3]
datas_scpandas = pd.DataFrame(datas_scale)
for i in (visu_sens_size):
    #sens = datas_pandas.iloc[i,:]
    sens_scale = datas_scpandas.iloc[:,i]
    sens_orig = datas_pandas.iloc[:,i]
    sens_orig.hist(bins=10)
    plt.show()
    sens_scale.hist(bins = 10)
    plt.show()

    plt.figure()
    plt.plot(sens_orig)
    plt.show()
    plt.figure()
    plt.plot(sens_scale)
    plt.show
```

## Creation of training set, validation set and test set

```python
from sklearn.model_selection import train_test_split
testset_perc = 0.3

#limiting data to have a smaller dataset to test
time_limit = 1000

X = datas_scale[:,:time_limit]
y = label[:,:time_limit]

X_train,X_test,y_train,y_test = train_test_split(X,y,stratify = label,random_state = 42,test_size = testset_perc)



print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```
